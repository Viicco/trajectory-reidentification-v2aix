```python
import matplotlib.pyplot as plt

# Données brutes des scores IDF1
idf1_scores = {
    "Naïf": 12.76,
    "Seuil": 22.53,
    "Prudent": 32.04,
    "NNPDA": 33.52,
    "Kalman": 46.15,
    "Graph": 28.72
}

# Trier les attaques par IDF1 croissant
idf1_sorted = dict(sorted(idf1_scores.items(), key=lambda x: x[1]))

# Couleurs (autant que nécessaire, ou cycle)
colors = plt.cm.tab10.colors[:len(idf1_sorted)]

# Création du plot
plt.figure(figsize=(10, 6))
bars = plt.bar(idf1_sorted.keys(), idf1_sorted.values(), color=colors)

# Ajouter les valeurs sur chaque barre
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f'{height:.1f}%', 
             ha='center', va='bottom', fontsize=10)

# Personnalisation
plt.ylabel("IDF1 Score (%)")
plt.title("IDF1 – Relinking Attacks on FCMZ (Ascending Order)")
plt.xticks()
plt.ylim(0, max(idf1_sorted.values()) + 10)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
